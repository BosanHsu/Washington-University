```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

```python
file_path = 'Team4.xlsx'
finance_df = pd.read_excel(file_path, sheet_name='Finance')
performance_df = pd.read_excel(file_path, sheet_name='Student_Performance')
```

```python
finance_df.head()
```

```python
performance_df.head()
```

```python
merged_df = pd.merge(performance_df, finance_df, on=['School_Name', 'Year'], how='inner')
merged_df.head()
```

```python
merged_df.dtypes
```

```python
columns_to_convert = [
    'Senior_High_Fund',
    'Instruction_and_Curriculum_Development_Services',
    'Instructional_Staff_Training_Services',
    'Professional_Development',
    'Instruction_Related_Technology',
    'Certificated_Regular_Salaries',
    'Certificated_Administrators_Salaries'
]
```

```python
def convert_to_float(column):
    return column.replace({'\$': '', ',': ''}, regex=True).astype(float)

for column in columns_to_convert:
    merged_df[column] = convert_to_float(merged_df[column])
    
merged_df.dtypes
```

```python
merged_df.isna().sum()
```

```python
# drop columns with more than 10 NAs
na_counts = merged_df.isna().sum()
columns_to_drop = na_counts[na_counts >= 10].index
cleaned_df = merged_df.drop(columns=columns_to_drop)
cleaned_df.describe()
```

```python
cleaned_df.isna().sum()
```

```python
# In Missouri, approximately 29% of the K-12 students are in high school
cleaned_df["high_school_students"] = cleaned_df["Total_Students"]*0.29 #
```

```python
financial_data = [
    'Instruction_and_Curriculum_Development_Services',
    'Instructional_Staff_Training_Services',
    'Professional_Development',
    'Instruction_Related_Technology',
    'Certificated_Regular_Salaries',
    'Certificated_Administrators_Salaries'
]
```

```python
cleaned_df["Senior_High_Fund"] = cleaned_df["Senior_High_Fund"]/cleaned_df["high_school_students"]

for column in financial_data:
    cleaned_df[column] = cleaned_df[column]/cleaned_df["Total_Students"]
```

```python
output_file_path = 'cleaned_data.xlsx'
cleaned_df.to_excel(output_file_path, index=False)
```

```python
cleaned_df.head()
```

```python
top_school_df = cleaned_df[cleaned_df["School_Name"] != "Bayless"]

potential_reasons = [
    'Diversity_Rate',
    'English_Learners_Rate',
    'Students_Eligible_for_Free_and_Reduced_Price_Lunch',
    'Senior_High_Fund',
    'Instruction_and_Curriculum_Development_Services',
    'Instructional_Staff_Training_Services',
    'Professional_Development',
    'Instruction_Related_Technology',
    'Certificated_Regular_Salaries',
    'Certificated_Administrators_Salaries'
]

performance_columns = [
    'Graduation_Rate_4_Year',
    '4yr_College',
    'ACT_Score'
]

correlation_matrix = top_school_df[potential_reasons + performance_columns].corr()
print(correlation_matrix.loc[potential_reasons, performance_columns])
```

```python
plt.figure(figsize=(12, 6))
sns.heatmap(correlation_matrix.loc[potential_reasons, performance_columns], annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix: Student Performance vs Financial Indicators')
plt.xlabel('Student Performance Indicators')
plt.ylabel('Financial and Other Indicators')
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.show()
```
